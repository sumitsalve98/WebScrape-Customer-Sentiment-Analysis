# Q4. Finding out what price range have most selling products

```python
def price_range(price):
    if price <= 25000:
        cate = '0-25000 ₹'
    elif price > 25000 and price <= 50000:
        cate = '25000-50000 ₹'
    elif price > 50000 and price <= 75000:
        cate = '50000-75000 ₹'
    elif price > 75000 and price <= 100000:
        cate = '75000-100000 ₹'
    elif price > 100000 and price <= 125000:
        cate = '100000-125000 ₹'
    elif price > 125000 and price <= 150000:
        cate = '125000-150000 ₹'
    elif price > 150000 and price <= 175000:
        cate = '150000-175000 ₹'
    elif price > 175000 and price <=200000:
        cate = '175000-200000 ₹'
    elif price > 200000:
        cate = '200000+ ₹'
    return cate
```

```python
df['price_range'] = df['price'].apply(price_range)
```

```python
df.head()
```

```python
df_price_range = df['sold'].groupby(df['price_range']).sum().to_frame().reset_index().sort_values(by='sold', ascending=False)
df_price_range.head()
```

```python
fig=px.bar(df_price_range, x='price_range' , y='sold', text='sold', color='price_range',
          title = "Price range and Total sales")
fig.update_layout(xaxis_title="price range", yaxis_title="Total sales")
fig.update_layout(legend=dict(title='Price Range'))
fig.update_traces(textposition="outside")

fig.write_image(r"C:\Users\sumit\End_To_End_DataAnalysis\images\Price range and Total sales.jpeg", width=1280, height=720, engine="kaleido", scale=2)
Image(fig.to_image(format="png", width=1280, height=720, engine="kaleido", scale=1))
```

```python
fig = px.pie(df_price_range, values='sold', names='price_range', title='% of sales by price range')
fig.write_image(r"C:\Users\sumit\End_To_End_DataAnalysis\images\percentage of sales by price range.jpeg", width=1280, height=720, engine="kaleido", scale=2)
Image(fig.to_image(format="png", width=1280, height=720, engine="kaleido", scale=1))
```

# Trend in sales by month

```python
# lets convert numbers 1 to 12 into months jan to dec

import calendar

def num_to_month(num):
    month_name = calendar.month_name[num]
    return month_name
```

```python
df['purchase_month']=df['purchase_month'].map(num_to_month)
```

```python
df_sales_month = df['sold'].groupby(df['purchase_month']).sum().reset_index()
months = pd.DataFrame({"purchase_month":['January','February','March','April','May','June','July','August','September','October','November','December']})
df_sales_month = months.merge(df_sales_month, on='purchase_month',how='left')
df_sales_month.head()
```

```python
fig = px.bar(df_sales_month, x='purchase_month', y='sold',text='sold', color='sold',
             color_continuous_scale="Viridis",
             title = "Trend in sales by month")
fig.update_layout(xaxis_title="Month", yaxis_title="Total sales")
fig.write_image(r"C:\Users\sumit\End_To_End_DataAnalysis\images\Trend in sales by month.jpeg", width=1280, height=720, engine="kaleido", scale=2)
Image(fig.to_image(format="png", width=1280, height=720, engine="kaleido", scale=1))
```

# 3. Geographical patterns in sales

### 3.2. indian states geojson file downloaded from here:  https://un-mapped.carto.com/tables/states_india/public/map

```python
statewise_sale = pd.DataFrame(df['state'].value_counts().reset_index())
statewise_sale.rename(columns={'index':'st_nm', 'state':'products_sold'}, inplace=True)
statewise_sale.head()
```

```python
import json
```

```python
path = r"D:\flipkart reviews\coordinates\states_india.geojson"
```

```python
indian_states = json.load(open(path, 'r'))
```

```python
indian_states['features'][0]['properties']
```

```python
import folium
import streamlit as st
from streamlit_folium import folium_static
```

```python
m = folium.Map(location=[23.47,72.94], tiles='CartoDB positron',name="Light Map",
              zoom_start=5,attr='My Data Attribution')
choice = ['products_sold']
choice_selected = st.selectbox("Select Choice ", choice)

folium.Choropleth(
    geo_data=indian_states,
    name="choropleth",
    data=statewise_sale,
    columns=["st_nm", choice_selected],
    key_on="feature.properties.st_nm",
    fill_color="YlOrRd",
    fill_opacity=0.7,
    line_opacity=.1,
    legend_name=choice_selected+"(%)",
).add_to(m)
folium.features.GeoJson(path, name="LSOA Code",
                           popup=folium.features.GeoJsonPopup(fields=['st_nm'])).add_to(m)
```
